```python
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: clean plot style
sns.set(style='whitegrid')

# Plot all numeric features
df.hist(figsize=(12, 8), color='lightgreen', edgecolor='black')
plt.suptitle("Feature Distributions", fontsize=16)
plt.show()
```

## Temp vs Crop

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='label', y='temperature', data=df)
plt.xticks(rotation=90)
plt.title("Crop vs Temperature")
plt.ylabel("Temperature (C)")
plt.show()
```

## Rain vs Crop

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='label', y='rainfall', data=df)
plt.xticks(rotation=90)
plt.title("Crop vs Rainfall")
plt.ylabel("Rainfall (mm)")
plt.show()
```

## ph vs Crop

```python
plt.figure(figsize=(12, 5))
sns.boxplot(x='label', y='ph', data=df)
plt.xticks(rotation=90)
plt.title("Crop vs Soil pH")
plt.ylabel("pH")
plt.show()
```

## Correlation

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Feature Correlation Heatmap")
plt.show()
```
